```python
class SmartDevices:
    def __init__(self, name, device_id):
        self.name = name
        if not device_id:
            raise ValueError("Device ID cannot be empty.")
        self.__device_id = device_id
        self.__power_status = False

    @property
    def device_id(self):
        return self.__device_id

    @property
    def power_status(self):
        return self.__power_status

    def _set_power_status(self, status: bool):
        self.__power_status = status

    def turn_on(self):
        self._set_power_status(True)
        print(f"[{self.name}] is now ON.")

    def turn_off(self):
        self._set_power_status(False)
        print(f"[{self.name}] is now OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Name: {self.name} | Device ID: {self.__device_id} | Power: {status}")

    # CHILD CLASS 1: SECURITY CAMERA
class SecurityCamera(SmartDevices):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__recording_status = False

    @property
    def recording_status(self):
        return self.__recording_status

    def start_recording(self):
        if not self.power_status:
            print(f"[{self.name}] cannot record while device is OFF. Turn it ON first.")
            return
        self.__recording_status = True
        print(f"[{self.name}] recording started.")

    def stop_recording(self):
        self.__recording_status = False
        print(f"[{self.name}] recording stopped.")

    def display_info(self):
        super().display_info()
        rec = "Recording" if self.__recording_status else "Not recording"
        print(f" Recording Status: {rec}")

    # CHILD CLASS 2: SMART LIGHT
class SmartLight(SmartDevices):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__brightness = 0

    @property
    def brightness(self):
        return self.__brightness
    @brightness.setter
    def brightness(self, value):
        if 0 <= value <= 100:
            self.__brightness = value
        else:
            print("Invalid brightness value. Must be between 0 and 100.")

    def increase_brightness(self, amount=10):
        new_value = self.__brightness + amount
        self.brightness = min(new_value, 100)
        print(f"[{self.name}] brightness increased to {self.__brightness}%.")

    def decrease_brightness(self, amount=10):
        new_value = self.__brightness - amount
        self.brightness = max(new_value, 0)
        print(f"[{self.name}] brightness decreased to {self.__brightness}%.")

    def display_info(self):
        super().display_info()
        print(f"  Brightness: {self.__brightness}%.")


    # CHILD CLASS 3: TEMPERATURE SENSOR
class TemperatureSensor(SmartDevices):
    def __init__(self, name, device_id, temperature=25.0):
        super().__init__(name, device_id)
        self.__temperature = temperature

    @property
    def temperature(self):
        return self.__temperature

    @temperature.setter
    def temperature(self, value):
        self.__temperature = value

    def read_temperature(self):
        print(f"[{self.name}] current temperature: {self.__temperature}^C")

    def display_info(self):
        super().display_info()
        print(f"  Temperature: {self.__temperature}^C")


    # MENU-DRIVEN INTERFACE
def choose_device(devices):
    print("\nSelect a device:")
    for index, device in enumerate(devices, start=1):
        print(f"{index}. {device.name}")
    try:
        choice = int(input("Enter device number:"))
        if 1 <= choice <= len(devices):
            return devices[choice - 1]
        else:
            print("Invalid selection.")
            return None
    except ValueError:
        print("Please enter a valid number.")
        return None

def main():
    camera = SecurityCamera("Front Door Camera", "CAM001")
    light = SmartLight("Living Room", "LGT001")
    sensor = TemperatureSensor("Kitchen Sensor", "TMP001")

    devices = [camera, light, sensor]

    while True:
        print("\n===== SMART DEVICE MANAGEMENT SYSTEM =====")
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start Recording")
        print("7. Exit")

        choice = input("Enter your choice (1-7):")

        if choice == "1":
            device = choose_device(devices)
            if device:
                device.display_play()

        elif choice == "2":
            device = choose_device(devices)
            if device:
                device.turn_on()

        elif choice == "3":
            device = choose_device(devices)
            if device:
                device.turn_off()

        elif choice =="4":
            if isinstance(sensor, TemperatureSensor):
                sensor.read_temperature()

        elif choice == "5":
            if not light.power_status:
                print(f"[{light.name}] turn the light on first.")
                continue
            direction = input("Increase or decrease brightness? (i/d):").lower()
            if direction == "i":
                light.increase_brightness()
            elif direction == "d":
                light.decrease_brightness()
            else:
                print("Invalid option.")

        elif choice == "6":
            camera.turn_on()
            camera.start_recording()

        elif choice == "7":
            print("Exiting Smart Device Management System. Goodbye!")
            break

        else:
            print("Invalid choice. Please select a number between 1 and 7.")


    # PROGRAM ENTRY POINT
if __name__ == "__main__":
    main()
```
